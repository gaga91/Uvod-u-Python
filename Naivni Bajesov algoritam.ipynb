U prvom primeru ćemo generisati sintetičke podatke koristeći scikit-learn, nad kojima cemo obučiti naivni Bajesov algoritam.

kreiramo skup podataka sa šest karakteristika, tri klase i 800 uzoraka koristeći funkciju `make_classification`.

```python
from sklearn.datasets import make_classification

X, y = make_classification(
    n_features=6,
    n_classes=3,
    n_samples=800,
    n_informative=2,
    random_state=1,
    n_clusters_per_class=1,
)
```

Koristićemo matplotlib.piplot funkciju `scatter` da vizualizujemo skup podataka.

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=y, marker="*");
```

Kao što možemo primetiti, postoje tri tipa ciljnih oznaka, pa ćemo trenirati model višeklasne klasifikacije.

Pre nego što započnemo proces obuke, potrebno je da podelimo skup podataka za obuku (trening skup) i test skup podataka.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=125
)
```

Napravite generički naivni Bajesov algoritam. 

```python
from sklearn.naive_bayes import GaussianNB

# Build a Gaussian Classifier
model = GaussianNB()

# Model training
model.fit(X_train, y_train)

# Predict Output
predicted = model.predict([X_test[6]])

print("Actual Value:", y_test[6])
print("Predicted Value:", predicted[0])  #Both actual and predicted values are the same. 
```

Evaluacija modela
Prvo ćemo predvideti vrednosti za skup podataka testa i koristiti ih za izračunavanje tačnosti i F1 rezultata.


```python

from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    #ConfusionMatrixDisplay,
    f1_score,
)

y_pred = model.predict(X_test)
accuray = accuracy_score(y_pred, y_test)
f1 = f1_score(y_pred, y_test, average="weighted")

print("Accuracy:", accuray)
print("F1 Score:", f1)
```
