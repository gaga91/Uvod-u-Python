# Sistematski uzorak

```python
#generisemo nasumicne podatke koji ce sluziti kao podaci o populaciji
# Generisacemo 10000 tacaka iz normalne raspodele, m=10, std=2

N=10000
mu=10
std=2
populacija_df=np.random.normal(mu, std, N)
```

```python
import pandas as pd
# 'korak' period u kom se svaki clan uzorka bira
    #'korak' izracunacemo ga tako sto ukupnu velicinu populacije podelimo sa zeljenom velicinom uzorka
def SistematskiUzorak(df,korak): 
    id=pd.Series(np.arange(1,len(df),1))
    df=pd.Series(df)
    df_pd=pd.concat([id,df],axis=1)
    df_pd.columns=["id", "data"]
    #vratice ravnomerno rasporedjene vrednosti
    selected_index=np.arange(8,len(df),korak)
    # iloc daje podatke na odredjenim indexima
    sitematskiUzorak= df_pd.iloc[selected_index]
    return(sitematskiUzorak)

    
```

```python
n=10 #obim uzorka
korak=int(N/n)
NasSistematskiUzorak=SistematskiUzorak(populacija_df,korak)
NasSistematskiUzorak
```

# Klasterovanje

Prvo, generišemo podatke koji će služiti kao podaci o populaciji sa 10.000 posmatranja, a ti podaci se sastoje od sledeće 4 varijable: <br>
Price: generisana korišćenjem uniformne distribucije na intervalu od 1 do 4, <br>
Id <br>
event_type: koja je kategorijska promenljiva sa 3 moguće vrednosti {tipe1, tipe2, tipe3} <br>
click: binarna promenljiva uzima vrednosti {0: nema klika, 1: klik} <br>

```python
#generisanje baze podataka 

price=pd.Series(np.random.uniform(1,4,size=N))
id=pd.Series(np.arange(0,len(price),1))
event_type=pd.Series(np.random.choice(["type1","type2","type3"],size=len(price)))
click=pd.Series(np.random.choice([0,1],size = len(price)))


df=pd.concat([id, price,event_type,click],axis=1)
df.columns=["id","price","type","click"]
df
```

```python
#funkcija koja izvlaci uzorak klasterovanje
#n_per_cluster broj elemenata u klasteru 
def getClusteredSample(df, n_per_cluster, num_selectCluster):
    
    N=len(df)
    K=int(N/n_per_cluster) #K nam oznacava broj klastera
    data=None
    for k in range(K): #malo k broj koji je klaster po redu; k=1,2,3...,K
        sample_k=df.sample(n_per_cluster) #Biramo n_per_cluster elemenata iz baze i stavljamo u k-ti klaster 
        #za svaki red ce pisati u kom je klasteru, pa dpdajemo kolomu cluster za svaki element koji se nlazsai u nasem klasteru
        sample_k["cluster"]=np.repeat(k, len(sample_k))
        #drop sklanja redove ili kolone
        #mi koristimo ovde drop da bi sklonili elemente koji su vec izabrani u klaster, da se ne bi ponavljali
        df=df.drop(index=sample_k.index)
        data=pd.concat([data, sample_k],axis=0)
    IzabraniKlasteri=np.random.randint(0,K,num_selectCluster) 
    uzorak=data[data.cluster.isin(IzabraniKlasteri)]
    return(uzorak)

NasUzorak=getClusteredSample(df=df,n_per_cluster =100, num_selectCluster=2)
NasUzorak
    
   
```

```python
def get_weighted_sample(df,n):
    def get_class_probabilty(x):
        #racunamo tezinu koju cemo da pridruzimo podacima koji su imali dogadjaj klika 1
        weight_x=int(np.rint(n*len(x[x.click!=0])/len(df[df.click!=0])))
        sampled_x=x.sample(weight_x).reset_index(drop=True)  
        return(sampled_x)
    weighted_sample=df.groupby('type').apply(get_class_probabilty)  
    print(weighted_sample["type"].value_counts())
    return(weighted_sample)
Uzorak=get_weighted_sample(df,100)
Uzorak                     
```

# Stratifikovan uzorak

```python
def get_startified_sample(df,n,num_clusters_needed):
    N = len(df)
    num_obs_per_cluster = int(N/n)
    K = int(N/num_obs_per_cluster)

    def get_weighted_sample(df,num_obs_per_cluster):
        def get_sample_per_class(x):
            n_x = int(np.rint(num_obs_per_cluster*len(x[x.click !=0])/len(df[df.click !=0])))
            sample_x = x.sample(n_x)
            return(sample_x)
        weighted_sample = df.groupby("type").apply(get_sample_per_class)
        return(weighted_sample)

    stratas = None
    for k in range(K):
        weighted_sample_k = get_weighted_sample(df,num_obs_per_cluster).reset_index(drop = True)
        weighted_sample_k["cluster"] = np.repeat(k,len(weighted_sample_k))
        stratas = pd.concat([stratas, weighted_sample_k],axis = 0)
        df.drop(index = weighted_sample_k.index)
    selected_strata_clusters = np.random.randint(0,K,size = num_clusters_needed)
    stratified_samples = stratas[stratas.cluster.isin(selected_strata_clusters)]
    return(stratified_samples)

sample = get_startified_sample(df = df,n = 100,num_clusters_needed = 2)
sample
```
